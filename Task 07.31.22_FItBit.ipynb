```python
# Other answer 
# Create a function to check healthy person with more than 1800 calories

def calorieCheck(a):
    if a>=1800:
        return 1
    else:
        return 0
```

```python
# Create a column to store healthy days values with more than 1800 calories

fit_df['HealthyDay'] = fit_df['Calories'].apply(calorieCheck)
```

```python
# len(fit_df[fit_df['healthy_days'] == 1])
```

```python
fit_df[['Id', 'HealthyDay', 'Calories']]
```

```python
healthy_df = fit_df.groupby(['Id'])[['HealthyDay', 'Calories']].sum()
```

```python
healthy_df.head()
```

```python
# Create a column to store total number of logged days per id

healthy_df['LoggedDays'] = fit_df.groupby('Id')['ActivityDate'].count().sort_values(ascending=False)
```

```python
# Create a new column to findout average of calorie burnt

healthy_df['AvgCalorie'] = healthy_df['Calories']/healthy_df['LoggedDays']
```

```python
# Create a column to findout the average health index, hera we devide healthy days by logged days.

healthy_df['AvgHealthIndex'] = healthy_df['HealthyDay']/healthy_df['LoggedDays']
```

```python
healthy_df.sort_values(by=['AvgCalorie', 'HealthyDay'], ascending=False).head()
```

```python
healthy_df.describe()
```

```python
# Classify the health status

def healthStatus(a,b):
    if ((a > 1800) or (b>0.8)):
        return 'Healthy'
    else:
        return 'Unhealthy'
```

```python
healthy_df['healthStatus'] =healthy_df.apply(lambda healthy_df: healthStatus(healthy_df['AvgCalorie'], healthy_df['AvgHealthIndex']), axis=1)
```
