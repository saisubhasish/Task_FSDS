```python
def dict_to_list(d):
    l=[]
    for i in d.values():
        if type(i)!=dict:
            l.append(i)
        if type(i)==dict:
            for j in i.values():
                l.append(j)
    return l
```

```python
d = {'k1' :"value" , "k2" : "values " ,"k3" : { "k12" : "sudh" , "k13"  : "gafasd"}}
dict_to_list(d)
```

```python
def concatenation(*l):
    s=''
    for x in l:
        if type(x)!=list:
            s+=x
        if type(x)==list:
            for y in x:
                s+=y
    return s
```
