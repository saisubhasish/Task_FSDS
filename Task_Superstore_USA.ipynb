```python
# 8 . Create a new coulmn and try to find out a diffrence between order date and shipment date

df_data1['Delivery Time(days)'] = df_data1['Ship Date'] - df_data1['Order Date']
type(df_data1['Delivery Time(days)'][0])
df_data1['Delivery Time(days)']
```

```python
# 9 . base on question number 8 find out for which order id we have shipment duration more than 10 days 

# Converting time delta format to str/object
df_data1['Delivery Time(days)'] = df_data1['Delivery Time(days)'].astype(str)

# Strip "days" from Delivery_Time column
df_data1['Delivery Time(days)'] = df_data1['Delivery Time(days)'].str.replace('days', "")

# Converting to int format to perform operations
df_data1['Delivery Time(days)'] = df_data1['Delivery Time(days)'].astype(int)

df_data1[df_data1['Delivery Time(days)'] > 10]
```

```python
# 10 . Try to find out a list of a returned order which sihpment duration was more then 15 days and find out that region manager as well 

# Inner join to find common orders
return_orders = pd.merge(df_data1, df_data2, how='inner')

# finding number of days
return_orders['Delivery_Time(days)'] = df_data1['Ship Date'] - df_data1['Order Date']

# converting column to str/obj type
return_orders['Delivery_Time(days)'] = return_orders['Delivery_Time(days)'].astype(str)

return_orders['Delivery_Time(days)']

# Removing the substring day
return_orders['Delivery_Time(days)'] = return_orders['Delivery_Time(days)'].str.replace('days','')

# Converting to int to perform operations
# Removing the substring day
return_orders['Delivery_Time(days)'] = return_orders['Delivery_Time(days)'].astype(int)

return_orders[return_orders['Delivery_Time(days)'] > 15]
```

```python
# 11 . Gorup by region and find out which region is more profitable 

max_profit_region = df_data1[['Region', 'Profit']].groupby('Region').sum().reset_index()
max_profit_region[max_profit_region['Profit'] == max(max_profit_region['Profit'])]
```

```python
# 12 . Try to find out overalll in which country we are giving more didscount 

max_discount_country = df_data1[['State or Province', 'Discount']].groupby('State or Province').sum().reset_index()
max_discount_country[max_discount_country['Discount'] == max(max_discount_country['Discount'])]
```

```python
# 13 . Give me a list of unique postal code 

postal_codes = []
for i in df_data1['Postal Code'].unique():
    postal_codes.append(i)
    
print(postal_codes)
```

```python
# 14 . which customer segement is more profitalble find it out . 

max_profit_customer_segment = df_data1[['Customer Segment', 'Profit']].groupby('Customer Segment').sum().reset_index()
max_profit_customer_segment[max_profit_customer_segment['Profit'] == max(max_profit_customer_segment['Profit'])]
```

```python
# 15 . try to find out the 10th most loss making product catagory .

category = df_data1[['Product Category', 'Profit']]
loss = category[category['Profit'] < 0].groupby('Product Category').sum().reset_index()
loss[loss['Profit'] == min(loss['Profit'])]
loss
```

```python
# 16 . Try to find out 10 top  product with highest margins 

a = df_data1[['Product Base Margin', 'Product Name']]
b = pd.DataFrame(a[a['Product Base Margin'] == max(a['Product Base Margin'])])
b
```
