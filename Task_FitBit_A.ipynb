```python
# Reading all the sheets in one go

df = pd.read_excel("Superstore_USA.xlsx", sheet_name= None)
```

```python
# We gor this dictionary with key as sheet name and DataFrame as it's value

df.keys()
# df['Orders']
```

```python
df_orders = df['Orders']
df_returns = df['Returns']
df_users = df['Users']
```

# 2 . while loading this data you dont have to create a table manually you can use any automated approach to create a table and load a data in bulk in table

```python
connection = conn.connect(host='localhost', user='root', passwd='Root_mysql')
cursor = connection.cursor()
cursor.execute("create database superstore;")
```

```python
mysql_engine = engine.create_engine("mysql+pymysql://root:Root_mysql@localhost/practice")
df_orders.to_sql("orders", index=False, con=mysql_engine)
df_returns.to_sql("returns", index=False, con=mysql_engine)
df_users.to_sql('users', index=False, con=mysql_engine)
```

# 3 . Find out how return that we ahve recived and with a product id 

```python
df_returns.merge(df_orders)[['Order ID', 'Product Name']]
```

# 4 . try  to join order and return data both in sql and pandas 

```python
# All the orers which got the return request are merged on the basis of 'Order ID' 
# because this is the only common column in both the dataframe

df_returns.merge(df_orders)
```

# 5 . Try to find out how many unique customer that we have 

```python
df_orders[['Customer Name']].drop_duplicates()
# df_orders[['Customer Name']].drop_duplicates(keep=False)
```

```python
len(df_orders[['Customer Name']].drop_duplicates())
```

# 6 . try to find out in how many regions we are selling a product and who is a manager for a respective region 

```python
df_orders.merge(df_users)[['Region', 'Manager']].drop_duplicates()
```

# 7 . find out how many different differnet shipement mode that we have and what is a percentage usablity of all the shipment mode with respect to dataset 

```python
df_orders['Ship Mode'].value_counts()/sum(df_orders['Ship Mode'].value_counts())*100
```

# 8 . Create a new coulmn and try to find our a diffrence between order date and shipment date

```python
df_orders['Shipment Duration'] = pd.to_timedelta(df_orders['Ship Date'] - df_orders['Order Date'])
```

```python
df_orders
```

# 9 . base on question number 8 find out for which order id we have shipment duration more than 10 days 

```python
df_orders[df_orders['Shipment Duration'] > pd.to_timedelta(10, unit='D')][['Order ID', 'Shipment Duration']]
```

# 10 . Try to find out a list of a returned order which sihpment duration was more then 15 days and find out that region manager as well 

```python
# There are no records with delay more than 15 days

df_orders.merge(df_returns)[['Shipment Duration']].drop_duplicates()
```

# To verify for 5 days

```python
df_orders[df_orders['Shipment Duration'] > pd.to_timedelta(5, unit='D')].merge(df_returns).merge(df_users)
```

# 11 . Gorup by region and find out which region is more profitable 

```python
df_orders.groupby('Region')['Profit'].sum().sort_values(ascending=False)
```

```python
df_orders.groupby('Region')['Profit'].sum().sort_values(ascending=False)[:1]
```

# 12 . Try to find out overalll in which country we are giving more discount 

```python
df_orders.groupby('State or Province')['Discount'].mean().sort_values(ascending=False).head()
```

```python
df_orders.groupby('State or Province')['Discount'].mean().sort_values(ascending=False).head()[:1]
```
